```python
def codifica(frase: str) -> str:
  S = ""

  for i in range(0, len(frase)):
    S = S + chr(ord(frase[i])+3)

  return S

codifica("Eu")
```

```python
def F1(nome: str) -> str:
  n = ""
  i = 0
  while i <len(nome):
    if i==0 or nome[i-1] == " ":
      n += nome[i] + "."
    i += 1
  return n

F1(' ')
```

```python
def gerar_emoticons():
  prefixo = ":-"

  for letra in ")(D*ox)":
    print(prefixo+letra)

gerar_emoticons()
```
